- с возрастом удовлетворенность возрастает. Самые неудовлетворенные - пассажиры ~25 лет, самые удовлетворенные ~ 45 лет
- с увеличением дистанции возрастает удовлетворенность и чем короче дистанция, тем больше неудовлетворенность
- увеличение качества предоставляемых дополнительных услуг увеличивает удовлетворенность.     
  сильно снижает удовлетворенность:
  - низкое качество wi-fi
  - отсутствие чистоты
  - низкое качество или отсутствие простоты онлайн бронирования
  - плохое качество еды или напитков или отсутствие их
  - низкая комфортность места для сиденья 
  
  сильно повышает удовлетворенность:
  - качество еды и напитков
  - online boarding
  - качественное развлечение на борту (видео, музыка и т.п.)
  - высокое качество обслуживания на борту
  - качественное обслуживание багажа

### cluster

<b>Расчитаем оптимальное количество кластеров</b>

```python
distance_mat = pdist(ds_sample.drop(['satisfaction'], axis=1))
```

```python
Z = hierarchy.linkage(distance_mat, 'ward')
```

```python
plt.figure(figsize=(16, 6))
dn =hierarchy.dendrogram(Z, truncate_mode='lastp', show_leaf_counts=True, show_contracted=True)
plt.axhline(y=145, c='gray');
```

```python
plt.figure(figsize=(16, 6))
dn =hierarchy.dendrogram(Z, truncate_mode='lastp', p=4, show_leaf_counts=True, show_contracted=True);
```

```python
max_d = 145
```

```python
clusters = hierarchy.fcluster(Z, max_d, criterion='distance')
```

<b>по дендограмме данные оптимально можно разделить на 4 кластера

```python
plt.figure(figsize=(10, 10))
plt.scatter(ds_tsne[:, 0], ds_tsne[:, 1], c=clusters, cmap='Set1');
```

<b>Метод локтя.     
    
C помощь "метода локтя" найдем оптимальное количество кластеров

```python
Z.shape
```

```python
plt.figure(figsize=(16, 6))
plt.plot(range(1, 16), Z[-15:, 2][::-1], marker='o');
```

<b>Оптимальное количество кластеров 2, 3, 4

```python
clusters2 = hierarchy.fcluster(Z, 2, criterion='maxclust')
clusters3 = hierarchy.fcluster(Z, 3, criterion='maxclust')
clusters5 = hierarchy.fcluster(Z, 5, criterion='maxclust')
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(121)
plt.scatter(ds_tsne[:, 0], ds_tsne[:, 1], c=clusters2, cmap='Set1')
plt.title('2 clusters')
plt.subplot(122)
plt.scatter(ds_tsne[:, 0], ds_tsne[:, 1], c=clusters3, cmap='Set1')
plt.title('3 clusters');
```

<b>Инерция

```python
inertia = []
k = range(2, 16)

for i in k:
    km = KMeans(n_clusters=i)
    km.fit(ds_sample.drop(['satisfaction'], axis=1))
    inertia.append(silhouette_score(ds_sample.drop(['satisfaction'], axis=1), km.labels_))
```

```python
plt.figure(figsize=(15, 6))
plt.plot(k, inertia, marker='o')
plt.xlabel('k')
plt.ylabel('inertia')
```

<b>по инерции оптимальное количество кластеров - 2

```python
plt.figure(figsize=(16, 8))
plt.subplot(121)
plt.scatter(ds_tsne[:, 0], ds_tsne[:, 1], c=ds_sample['satisfaction'], cmap='Set1')
plt.title('cluster_satisfaction')
plt.subplot(122)
plt.scatter(ds_tsne[:, 0], ds_tsne[:, 1], c=clusters2, cmap='Set1_r')
plt.title('cluster_2');
```

<b>на картинке представленны данные, разделенные по меткам satisfaction (слева) и при помощи алгоритма (справа). Визуально видно, что разделения схожи.
