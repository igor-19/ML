```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
```

```python
def quality_report(prediction, actual):
    print('Accuracy: {:.3f}\nPrecision {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}'.format(
        accuracy_score(prediction, actual), 
        precision_score(prediction, actual), 
        recall_score(prediction, actual), 
        f1_score(prediction, actual)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(heart_df.drop(['TenYearCHD'], axis=1), heart_df['TenYearCHD'],
                                                   test_size=0.2, random_state=42, 
                                                    stratify=heart_df['TenYearCHD'])
```

### Scaler

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train_scaled, y_train)
```

```python
print("Train quality:")
quality_report(log_reg.predict(X_train_scaled), y_train)
print('\nTest quality:')
quality_report(log_reg.predict(X_test_scaled), y_test)
```

```python
featureImportance = pd.DataFrame({'feature': heart_df.drop(['TenYearCHD'], axis=1).columns,
                                 'importance': log_reg.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(['importance'], ascending=False, inplace=True)
featureImportance.plot(kind='bar', figsize=(15, 5));
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
def plot_roc_curve(prob_prediction, actual):
    fpr, tpr, thresholds = roc_curve(y_test, prob_prediction)
    auc_score = roc_auc_score(y_test, prob_prediction)
    
    plt.plot(fpr, tpr, label='ROC_curve')
    plt.xlabel('false positive rate')
    plt.ylabel('true positive rate')
    plt.plot([0.0, 1.0], [0.0, 1.0])
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.title('ROC AUC: {:.3f}'.format(auc_score))
    plt.show()
```

```python
fpr, tpr, threshold = roc_curve(y_test, log_reg.predict_proba(X_test_scaled)[:, 1])
```

```python
fpr[:10], tpr[:10], threshold[:10]
```

```python
plot_roc_curve(log_reg.predict_proba(X_test_scaled)[:, 1], y_test)
```

```python
log_reg_cv = LogisticRegressionCV(Cs=100)
```

```python
log_reg_cv.fit(X_train_scaled, y_train)
```

```python
print('Train quality:')
quality_report(log_reg_cv.predict(X_train_scaled), y_train)
print('\nTest quality:')
quality_report(log_reg_cv.predict(X_test_scaled), y_test)
```

```python
plot_roc_curve(log_reg_cv.predict_proba(X_test_scaled)[:, 1], y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, log_reg_cv.predict_proba(X_test_scaled)[:, 1] > 0.2)
conf_matrix = pd.DataFrame(data=cm, columns=['Predicted:0', 'Predicted:1'], index=['Actual:0', 'Actual:1'])
plt.figure(figsize=(5, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='YlGnBu', cbar=False);
```

### Эконометрический подход

```python
import statsmodels.api as am
```

```python
X_train_scaled_econ = pd.DataFrame(X_train_scaled, columns=heart_df.columns[:-1])
```

```python
y_train.shape, X_train_scaled.shape, y_test.shape, X_test_scaled.shape
```

```python
X_train_scaled_econ.head(3)
```

```python
model = am.Logit(y_train.values, X_train_scaled_econ)
result = model.fit()
result.summary()
```

-forward elimination (строим модель с одного фактора и добавляем факторы до тех пор, пока они значимы)   
-backward elimination (строим полную моель, находим pvalue, который максимален, удаляем этот фактор и т.д. находя оптимальную модель)

```python
np.argmax(result.pvalues), result.pvalues.max(), result.pvalues[7], X_train_scaled_econ.columns[7]
```

```python
result.pvalues
```

```python
col = X_train_scaled_econ.columns
col, len(col)
```

```python
for i in range(len(col)):
    model = sm.Logit(y_train.values, X_train_scaled_econ[col])
    result = model.fit()
    
    potentially_bad = col[np.argmax(result.pvalues)]
    
    if result.pvalues.max() > 0.05:
        col = [c for c in col if c != potentially_bad]
```
