```python
def outlier_detect_IQR(data, col, threshold=3):
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    upper_fence = data[col].quantile(0.75) + (IQR * threshold)
    lower_fence = data[col].quantile(0.25) - (IQR * threshold)
    
    data_free_outlier = data[col][data[col] < upper_fence]
    data_free_outlier = data_free_outlier[data_free_outlier > lower_fence]
        
    return data_free_outlier
```

```python
data_free_outlier = outlier_detect_IQR(data=ds_encoding, col='hours_per_week', threshold=1.5)
```

```python
def outlier_detect_whisker(data, col, threshold=1.5):
    IQR = data[col].quantile(0.75) - data[col].quantile(0.25)
    upper_fence = data[col].quantile(0.75) + (IQR * threshold)
    lower_fence = data[col].quantile(0.25) - (IQR * threshold)
    
    data_copy = data[col].copy(deep=True)
    index_upper = data[col][data[col] > upper_fence].index
    index_lower = data[col][data[col] < lower_fence].index
    
    data_copy.loc[index_upper] = upper_fence
    data_copy.loc[index_lower] = lower_fence
    
    return data_copy
```

```python
data_outlier_whisker = outlier_detect_whisker(data=ds_encoding, col='hours_per_week', threshold=1.5)
```

```python
plt.figure(figsize=(15, 3))
plt.subplot(1, 2, 1)
data_outlier_whisker.hist()
plt.subplot(1, 2, 2)
sns.boxplot(data_outlier_whisker);
```

```python
ds_encoding.hours_per_week.shape, data_outlier_whisker.shape
```

```python
### найдем feature в которых приобладают константные значения
```

```python
def constant_feature_detect(data, threshold=0.9):
    data_copy = data.copy(deep=True)
    constant_feature = {}
    for feature in data_copy.columns:
        predominant = (data_copy[feature].value_counts() / np.float(
            len(data_copy))).sort_values(ascending=False).values[0]
        if predominant >= threshold:
            constant_feature[feature] = predominant
    return constant_feature
```

```python
constant_features = constant_feature_detect(data=ds, threshold=0.5)
```
