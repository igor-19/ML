```python
df_categor = pd.get_dummies(df['code'], drop_first=True, prefix='code')
```

```python
df = pd.concat([df.drop('code', axis=1), df_categor], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.3)
```

```python
df.shape, df_train.shape, df_test.shape, df_train.shape[0] + df_test.shape[0]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df_train[['totsp', 'livesp', 'kitsp', 'dist', 'metrdist']])
```

```python
df_train_scale = scaler.transform(df_train[['totsp', 'livesp', 'kitsp', 'dist', 'metrdist']])
df_test_scale = scaler.transform(df_test[['totsp', 'livesp', 'kitsp', 'dist', 'metrdist']])
```

```python
df_train[['totsp', 'livesp', 'kitsp', 'dist', 'metrdist']] = df_train_scale
df_test[['totsp', 'livesp', 'kitsp', 'dist', 'metrdist']] = df_test_scale
```

```python
df_train.head(3)
```

```python
df_train.describe()
```

```python
y_train = df_train.price
y_test = df_test.price
```

```python
X_train = df_train.drop('price', axis=1).values
X_test = df_test.drop('price', axis=1).values
```

## Константный прогноз (base line)

1. предположим, что стоимость квартиры -> среднее значение. Сравним с этим предположением более сложные модели

```python
y_mean = np.mean(y_train)  # наивный прогноз
y_pred_naive = np.ones(len(X_test)) * y_mean
```

```python
y_pred_naive[:5]
```

## Метрики качества для регресии

```python
from sklearn import metrics
```

### MAE (mean absolute error)

средняя сумма, на которую ошиблись

```python
metrics.mean_absolute_error(y_test, y_pred_naive)
```

```python
metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred_naive))
```

##### т.к. цена прологорифмирована, необходимо взять экспоненту, что бы узнать предсказанную цену

```python
np.exp(y_mean)
```

### MSE (mean squared error)

```python
metrics.mean_squared_error(y_test, y_pred_naive)
```

```python
metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred_naive))
```

### RMSE  (root mean squared error)

```python
np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred_naive)))
```

### MAPE  (mean absolute percentage error)

```python
def mean_absolute_percentage_error(y_true, y_predict):
    y_true, y_pred = np.array(y_true), np.array(y_predict)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100 
```

```python
mean_absolute_percentage_error(y_test, y_pred_naive)
```

##### средняя ошибка примерно 5%

###  $R^2$  - коэффициент детерминации

```python
metrics.r2_score(y_test, y_pred_naive)
```

```python
def print_metrics(y_test, y_pred):
    print('MAE: ', metrics.mean_absolute_error(np.exp(y_test), np.exp(y_pred)))
    print('RMSE: ', np.sqrt(metrics.mean_squared_error(np.exp(y_test), np.exp(y_pred))))
    print('R2: ', metrics.r2_score(np.exp(y_test), np.exp(y_pred)))
    print('MAPE: ', mean_absolute_percentage_error(y_test, y_pred))
```

```python
print_metrics(y_test, y_pred_naive)
```

### Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred_regr = model.predict(X_test)
```

```python
# наивный прогноз
print_metrics(y_test, y_pred_naive)
```

```python
# регрессия
print_metrics(y_test, y_pred_regr)
```

##### вклад признаков в модель

```python
featureImportance = pd.DataFrame({'feature':df.drop('price', axis=1).columns, 'importance': model.coef_})
featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(['importance'], ascending=False, inplace=True)
featureImportance.plot(kind='bar', figsize=(10, 6))
plt.show()
```

### Lasso

```python
from sklearn.linear_model import Lasso
```

```python
model_Lasso = Lasso()
model_Lasso.fit(X_train, y_train)
y_pred_lasso = model_Lasso.predict(X_test)
```

```python
# наивный прогноз
print_metrics(y_test, y_pred_lasso)
```

```python
# lasso прогноз
print_metrics(y_test, y_pred_lasso)
```

```python
def feature_importance(model):
    featureImportance = pd.DataFrame({'feature': df.drop('price', axis=1).columns, 
                                  'importance': model.coef_})
    featureImportance.set_index('feature', inplace=True)
    featureImportance.sort_values(['importance'], ascending=False, inplace=True)
    featureImportance['importance'].plot(kind='bar');
```

```python
feature_importance(model_Lasso)
```

```python
model_Lasso.coef_
```

### GridSearch Lasso

```python
from sklearn.model_selection import GridSearchCV
```

```python
alpha = np.linspace(0.0001, 0.3, 10)
param_grid = {'alpha': alpha}
```

```python
model_Lasso = Lasso()
```

```python
grid_cv_lasso = GridSearchCV(model_Lasso, param_grid=param_grid, cv=5)
```

```python
grid_cv_lasso.fit(X_train, y_train)
```

```python
print('Лучшие значения параметра ', grid_cv_lasso.best_params_)
```

```python
mae_all = []

for a in alpha:
    model_Lasso_t = Lasso(alpha=a)
    model_Lasso_t.fit(X_train, y_train)
    y_pred_t = model_Lasso_t.predict(X_test)
    mae = metrics.mean_absolute_error(y_test, y_pred_t)
    mae_all.append(mae)
    print(f'alpha = {a}, mae = {mae}')
```

```python
mae_all
```

```python
plt.plot(alpha, mae_all);
```

```python
feature_importance(grid_cv_lasso.best_estimator_)
```

```python
grid_cv_lasso.best_estimator_
```

```python
print_metrics(y_test, grid_cv_lasso.best_estimator_.predict(X_test))
```
